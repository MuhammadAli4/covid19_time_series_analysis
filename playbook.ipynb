# TIME SERIES

```python
#TIME SERIES CONFIRMED

def time_series_confirmed(col): 
    series_confirmed = confirmed.loc[:, confirmed.columns[-col:]]
    region_name_confirmed= confirmed.loc[:,confirmed.columns[1]]
    pd.DataFrame(region_name_confirmed)
    df = pd.concat([series_confirmed, region_name_confirmed], axis=1, join="inner")
    time_series_per_country_confirmed = df.groupby(['Country/Region']).sum()  # same countries ki series values are summed
    
    return time_series_per_country_confirmed

```

```python
#TIME SERIES DEATHS

def time_series_deaths(col):

    series_deaths = deaths.loc[:, deaths.columns[-col :]]
    region_name_deaths= deaths.loc[:,deaths.columns[1]]
    pd.DataFrame(region_name_deaths)
    df = pd.concat([series_deaths, region_name_deaths], axis=1, join="inner")
    time_series_per_country_deaths = df.groupby(['Country/Region']).sum()
    
    return time_series_per_country_deaths

    
```

```python
#TIME SERIES RECOVERED

def time_series_recovered(col):
    
    series_recovered = recovered.loc[:, recovered .columns[-col:]]
    region_name_recovered = recovered.loc[:,recovered.columns[1]]
    pd.DataFrame(region_name_recovered )
    df = pd.concat([series_recovered , region_name_recovered], axis=1, join="inner")
    time_series_per_country_recovered  = df.groupby(['Country/Region']).sum()
    
    return time_series_per_country_recovered

```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split


def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   # not shown in the book
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)              # not shown
```

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
plt.axis([0, 40, -1, 1])                         # not shown in the book
plt.show()   
```
